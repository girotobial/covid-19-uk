```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import pandas as pd
import seaborn as sns 
from pathlib import Path
```

```python
def divide(num, denum, err):
    if denum > 0:
        return num / denum
    else:
        
        return err

def growth_ratio(iterable):
    growth_list = []
    for i, val in enumerate(iterable):
        if i == 0:
            ratio = 1

        if i > 0:
            ratio = divide(val, iterable[i - 1], 1)

        growth_list.append(ratio)
    return growth_list
```

```python
sns.set(style='ticks', context='talk')
```

```python
path = Path().cwd().parent
dailies = pd.read_csv(path / 'data' / 'DailyConfirmedCasesWithFeatures.csv', index_col='DateVal')
dailies
```

```python
dailies.plot(
    y='CumCases',
    figsize=(10, 6),
    marker='o',
);
_, top = plt.ylim()
plt.ylim(0, top)
plt.xlabel('Date')
plt.ylabel('Cumulative Cases')
plt.title('Cases of COVID-19 in the UK');
plt.xticks(rotation=45, ha='right');
plt.grid(which='major', axis='y');

plt.tight_layout();
sns.despine(left=True);
```

```python
dailies.plot(y='GrowthFactor', figsize=(10, 6));
left, right = plt.xlim()
plt.hlines(1, left, right, ls='--', color='k')
plt.xticks(rotation=45, ha='right');
plt.grid(which='major', axis='y');
sns.despine(left=True);
```

```python
weeklies = dailies[['Week', 'CMODateCount']].groupby('Week').sum()
weeklies['CumCases'] = weeklies['CMODateCount'].cumsum()
weeklies['GrowthFactor'] = growth_ratio(weeklies['CMODateCount'].values)
weeklies
```

```python
weeklies.plot(
    y='GrowthFactor',
    figsize=(12, 8)
);
plt.xlabel('Week');
plt.ylabel('Growth Factor');
left, right = plt.xlim()
plt.hlines(1, left, right, ls='--', color='k')
plt.title('Growth Factor of COVID-19 in the UK');
plt.xticks(rotation=45, ha='right');
plt.grid(which='major', axis='y');
sns.despine(left=True)
```
